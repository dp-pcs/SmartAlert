# 🔬 Model Bakeoff with TF-IDF Embeddings
This notebook uses the updated V4 dataset and TF-IDF features from the `message` column
to compare RandomForest, XGBoost, and LightGBM over multiple injected training rounds.

```python
from injection_harness_v4 import run_bakeoff_with_tfidf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
# Run the model injection + evaluation bakeoff
results = run_bakeoff_with_tfidf("data/splunk_logs_incidents_v4.csv", batch_size=10000, num_batches=5)
results
```

```python
# Plot F1 score trends by model over batch rounds
plt.figure(figsize=(10,6))
sns.lineplot(data=results, x="batch", y="f1", hue="model", marker="o")
plt.title("Model F1 Score Over Batches")
plt.ylabel("F1 Score")
plt.xlabel("Batch Number")
plt.xticks(range(1, 6))
plt.legend(title="Model")
plt.tight_layout()
plt.show()
```
