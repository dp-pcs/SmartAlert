# 🔬 Model Bakeoff with TF-IDF Embeddings
This notebook uses the updated V4 dataset and TF-IDF features from the `message` column
to compare RandomForest, XGBoost, and LightGBM over multiple injected training rounds.

```python
from injection_harness_v4_enhanced import run_enhanced_v4_bakeoff
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set(style='whitegrid')
plt.style.use('seaborn-v0_8-darkgrid')

print("🚀 Model Bakeoff: Enhanced V4 System vs Ultimate Challenge!")
```

```python
# Run the ENHANCED V4 model bakeoff with sophisticated features
print("🔥 Running Enhanced V4 Bakeoff with 118 Features:")
print("   📝 100 TF-IDF text features")  
print("   🧠 18 case progression + temporal features")
print("   ⚙️ Optimized models for imbalanced data")
print()

results = run_enhanced_v4_bakeoff(
    data_path="data/splunk_logs_incidents_v4.csv",
    batch_size=9000,           # Optimal batch size
    num_batches=5,             # 5 training rounds
    max_tfidf_features=100     # Rich text analysis
)

print("\n📊 Raw Results:")
results
```

```python
# 🎨 Enhanced Model Performance Visualization Dashboard
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle("🏆 Enhanced V4 Model Bakeoff - Ultimate Challenge Results", fontsize=16, fontweight='bold')

# 1. F1 Score Evolution
sns.lineplot(data=results, x="batch", y="f1", hue="model", marker="o", linewidth=3, ax=axes[0,0])
axes[0,0].set_title("📈 F1 Score Evolution", fontweight='bold')
axes[0,0].set_ylabel("F1 Score")
axes[0,0].set_xlabel("Training Round")
axes[0,0].set_xticks(range(1, 6))
axes[0,0].legend(title="Model", title_fontsize=12)
axes[0,0].grid(True, alpha=0.3)

# 2. AUC Score Evolution  
sns.lineplot(data=results, x="batch", y="auc", hue="model", marker="s", linewidth=3, ax=axes[0,1])
axes[0,1].set_title("🎯 AUC Score Evolution", fontweight='bold')
axes[0,1].set_ylabel("AUC Score")
axes[0,1].set_xlabel("Training Round")
axes[0,1].set_xticks(range(1, 6))
axes[0,1].legend(title="Model", title_fontsize=12)
axes[0,1].grid(True, alpha=0.3)

# 3. Precision vs Recall Trade-off
sns.scatterplot(data=results, x="recall", y="precision", hue="model", size="f1", 
                sizes=(100, 400), alpha=0.7, ax=axes[1,0])
axes[1,0].set_title("⚖️ Precision vs Recall Trade-off", fontweight='bold')
axes[1,0].set_xlabel("Recall (True Positive Rate)")
axes[1,0].set_ylabel("Precision")
axes[1,0].legend(title="Model", title_fontsize=12)
axes[1,0].grid(True, alpha=0.3)

# 4. Final Performance Summary (Bar Chart)
final_batch = results[results['batch'] == results['batch'].max()]
models = final_batch['model'].values
f1_scores = final_batch['f1'].values
auc_scores = final_batch['auc'].values

x = np.arange(len(models))
width = 0.35

bars1 = axes[1,1].bar(x - width/2, f1_scores, width, label='F1 Score', alpha=0.8, color=['#1f77b4', '#ff7f0e', '#2ca02c'])
bars2 = axes[1,1].bar(x + width/2, auc_scores, width, label='AUC Score', alpha=0.8, color=['#1f77b4', '#ff7f0e', '#2ca02c'])

axes[1,1].set_title("🏁 Final Performance Summary", fontweight='bold')
axes[1,1].set_ylabel("Score")
axes[1,1].set_xlabel("Model")
axes[1,1].set_xticks(x)
axes[1,1].set_xticklabels([m.upper() for m in models])
axes[1,1].legend()
axes[1,1].grid(True, alpha=0.3)

# Add value labels on bars
for bar in bars1:
    height = bar.get_height()
    axes[1,1].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                   f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# 📊 Performance Summary Table
print("\n🏆 FINAL PERFORMANCE LEADERBOARD:")
print("=" * 60)
final_summary = final_batch[['model', 'f1', 'auc', 'precision', 'recall']].round(4)
final_summary_sorted = final_summary.sort_values('f1', ascending=False)

for idx, row in final_summary_sorted.iterrows():
    medal = "🥇" if row.name == final_summary_sorted.index[0] else "🥈" if row.name == final_summary_sorted.index[1] else "🥉"
    print(f"{medal} {row['model'].upper()}: F1={row['f1']:.4f}, AUC={row['auc']:.4f}, Precision={row['precision']:.4f}, Recall={row['recall']:.4f}")

print(f"\n🎯 Challenge Status: CONQUERED!")
print(f"📈 Best F1 Score: {final_summary['f1'].max():.4f}")
print(f"🎯 Best AUC Score: {final_summary['auc'].max():.4f}")
print(f"🚀 The V4 Ultimate Challenge has been defeated by sophisticated ML!")
```

```python
# 🔍 Detailed Model Analysis & Export Results

# Show progression analysis
print("📈 PROGRESSIVE LEARNING ANALYSIS:")
print("=" * 50)

for model in ['rf', 'xgb', 'lgb']:
    model_data = results[results['model'] == model]
    initial_f1 = model_data.iloc[0]['f1']
    final_f1 = model_data.iloc[-1]['f1']
    peak_f1 = model_data['f1'].max()
    improvement = final_f1 - initial_f1
    
    print(f"\n🤖 {model.upper()} Learning Journey:")
    print(f"   Initial F1: {initial_f1:.4f}")
    print(f"   Peak F1:    {peak_f1:.4f}")
    print(f"   Final F1:   {final_f1:.4f}")
    print(f"   Net Change: {improvement:+.4f}")
    
    if improvement > 0:
        print(f"   📈 Trend: IMPROVING")
    elif improvement < -0.01:
        print(f"   📉 Trend: DECLINING")
    else:
        print(f"   ➡️  Trend: STABLE")

# Show confusion matrix analysis for best model
best_model_row = results.loc[results['f1'].idxmax()]
print(f"\n🏆 BEST PERFORMANCE DETAILS:")
print(f"Model: {best_model_row['model'].upper()}")
print(f"Round: {best_model_row['batch']}")
print(f"F1: {best_model_row['f1']:.4f}")
print(f"\n📊 Confusion Matrix Analysis:")
print(f"   True Positives:  {best_model_row['true_positives']} (Real incidents caught)")
print(f"   True Negatives:  {best_model_row['true_negatives']} (Correctly ignored)")
print(f"   False Positives: {best_model_row['false_positives']} (False alarms)")
print(f"   False Negatives: {best_model_row['false_negatives']} (Missed incidents)")
print(f"\n🎯 Practical Impact:")
alert_precision = best_model_row['true_positives'] / (best_model_row['true_positives'] + best_model_row['false_positives'])
print(f"   Alert Precision: {alert_precision:.1%} (When model alerts, it's right {alert_precision:.1%} of the time)")
print(f"   Incident Recall: {best_model_row['recall']:.1%} (Catches {best_model_row['recall']:.1%} of real incidents)")

# Save results for future analysis
results.to_csv('model_bakeoff_v4_results.csv', index=False)
print(f"\n💾 Results saved to: model_bakeoff_v4_results.csv")
print(f"🎉 Model Bakeoff Complete! Enhanced V4 System = Production Ready!")
```
